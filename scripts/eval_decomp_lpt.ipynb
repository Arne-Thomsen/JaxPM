```python
pdims=(2 ,2)
init_field_slices = []
field_slices = []

for i in range(pdims[0]):
    row_init_field = []
    row_field = []
    for j in range(pdims[1]):
        slice_index = i + j * pdims[0]
        row_field.append(np.load(f'field_{slice_index}.npy'))
        row_init_field.append(np.load(f'initial_conditions_{slice_index}.npy'))

    field_slices.append(np.hstack(row_field))
    init_field_slices.append(np.hstack(row_init_field))

field = np.vstack(field_slices)
initial_conditions = np.vstack(init_field_slices)
    
```

```python
#initial_conditions = np.load(f'initial_conditions_gathered.npy')
#field = np.load(f'field_gathered.npy')

figure(figsize=(12,6))
proj_axis = 2
box_size=200
subplot(121)
imshow(initial_conditions.sum(axis=proj_axis), cmap='magma', extent=[0, box_size, 0, box_size])
xlabel('Mpc/h')
ylabel('Mpc/h')
title('Initial conditions')
subplot(122)
imshow(field.sum(axis=proj_axis), cmap='magma', extent=[0,box_size,0,box_size])
xlabel('Mpc/h')
ylabel('Mpc/h')
title('1-LPT density field at z=0')
```
